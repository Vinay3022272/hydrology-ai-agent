```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(
    web_paths=("https://www.sarvam.ai/",),
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            class_=("post-title", "post-header")
        )
    ),
)

docs = loader.load()
all_docs.append(docs)

docs
```

```python
from __future__ import annotations

from pathlib import Path
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
def load_markdown_file(file_path: str):
    """
    Load a single markdown file using LangChain TextLoader.
    """
    loader = TextLoader(file_path, encoding="utf-8")
    docs = loader.load()

    # add custom metadata
    for doc in docs:
        doc.metadata["source_type"] = "markdown"
        doc.metadata["file_name"] = Path(file_path).name

    return docs
```
