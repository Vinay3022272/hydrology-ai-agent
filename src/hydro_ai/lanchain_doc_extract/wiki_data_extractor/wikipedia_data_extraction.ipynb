```python
import os
from langchain_community.utilities import WikipediaAPIWrapper
import markdown
```

```python
# CONFIG
OUTPUT_DIR = "generated_knowledge"
os.makedirs(OUTPUT_DIR, exist_ok=True)

OUTPUT_MD = os.path.join(OUTPUT_DIR, "rainfall_knowledge.md")
```

```python
import json
from pathlib import Path

with open(Path("rainfall_question.json"), "r", encoding="utf-8") as f:
    questions = json.load(f)
    
len(questions)    
```

```python
# WIKIPEDIA SETUP
wiki = WikipediaAPIWrapper(
    top_k_results=1,
    doc_content_chars_max=2000  # limit size (important)
)
```

```python
def clean_text(text):
    # remove unwanted sections
    text = text.split("References")[0]
    text = text.split("External links")[0]
    text = text.split("See also")[0]

    # clean formatting
    text = text.replace("\n", " ")
    text = " ".join(text.split())  # remove extra spaces

    return text.strip()


# BUILD MARKDOWN
markdown_content = "# Rainfall prediction\n\n"

for i, question in enumerate(questions):
    print(f"Fetching: {question}")

    try:
        raw_text = wiki.run(question)
        cleaned_text = clean_text(raw_text)

        # Markdown structure
        markdown_content += f"## {question}\n\n"
        markdown_content += f"{cleaned_text}\n\n"
        markdown_content += "---\n\n"

    except Exception as e:
        print(f"Error fetching {question}: {e}")

# SAVE MARKDOWN FILE
with open(OUTPUT_MD, "w", encoding="utf-8") as f:
    f.write(markdown_content)

print(" Markdown knowledge base created!")
```
