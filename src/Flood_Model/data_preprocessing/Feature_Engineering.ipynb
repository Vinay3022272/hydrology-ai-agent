### Why Convert LULC to Category Type
- Treats `lulc` as a nominal categorical variable, not as a continuous numeric value.
- Prevents misleading ordinal interpretation of land cover labels.
- Reduces memory usage compared to plain object/string storage.
- Keeps preprocessing simple and consistent before model-specific encoding.
- Allows flexible downstream handling: native categorical support (if available) or later encoding only when required.

###### One-hot encoding provides explicit separation of categories, while categorical encoding allows tree-based models to internally learn optimal splits without increasing feature dimensions

```python
df["lulc"] = df["lulc"].astype("category")
```
